```python
stopwordList = list()
with open('./stopword.txt', 'r', encoding='utf-8') as f:
    stopwordList = [i.replace('\n', '') for i in f.readlines()]
stopwordList
```

```python
wordCount = dict()
for w in cutList:
    if w not in stopwordList and len(w) > 1:
        if w in wordCount:
            wordCount[w] += 1
        else:
            wordCount[w] = 1
wordCount
```

```python
orderList = [(i, wordCount[i]) for i in wordCount]
orderList
```
