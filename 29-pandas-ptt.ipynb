```python
import pandas as pd
import glob
```

```python
glob.glob('res_gossiping/*.txt')
```

```python
rowData = list()
for article in glob.glob('res_gossiping/*.txt'):
    try:
        with open(article, 'r', encoding='utf-8') as f:
            tmpList = f.read().split('---split---')[1].split('\n')[1:-1]
            rowData.append(tmpList)
    except:
        pass
rowData
```

```python
columns = [r.split(': ')[0] for r in rowData[0]]
columns
```

```python
df = pd.DataFrame(data=rowData, columns=columns)
df
```

```python
def columns_filter(s):
    return s.split(': ')[-1]
```

```python
df2 = df.copy()
```

```python
df2['推']
```

```python
df2['推'].apply(columns_filter)
```

```python
df2
```

```python
for c in df2:
    print(c)
```

```python
for c in df2:
    df2[c] = df2[c].apply(columns_filter)
df2
```

```python
df2
```

```python
df3 = df.copy()
df3
```

```python
for c in df3:
    df3[c] = df3[c].apply(lambda x: x.split(': ')[-1])
df3
```

```python
df3.to_csv('ptt_article.csv', index=False, encoding='utf-8-sig')
```
